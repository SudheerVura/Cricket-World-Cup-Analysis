```python
# we have to combine all the arrays in to single array by appending

with open(r"C:\Users\asus\Downloads\data-analytics-project-for-beginners\t20_json_files\t20_json_files\t20_wc_batting_summary.json") as F1:
    data = json.load(F1)
    all_records = []
    
    for records in data:
        all_records.extend(records['battingSummary'])
        
df_batting = pd.DataFrame(all_records)
df_batting.tail()
```

```python
df_batting.head(10)
```

```python
# creating a new column for OUT/NOT OUT using dismissal column
# here lambda of x is fetching each value of that column
# using the apply function we are changing it


df_batting["out/not_out"] = df_batting.dismissal.apply(lambda x : "out" if len(x) > 0 else "not_out")
#df_batting.head(10)
df_batting.tail(10)
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€',''))
df_batting['batsmanName'].head(11)
```
